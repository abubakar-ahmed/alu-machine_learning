```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('fuel_econ.csv')
data.head(7)
```

# Interpretation:
# The histogram shows how fuel efficiency is distributed across vehicles.
# Most vehicles have a fuel efficiency between 15 and 25 miles per gallon (mpg),
# indicating that a large portion of the dataset consists of mid-range efficiency vehicles.
# Few vehicles have extreme values for fuel efficiency (either very low or very high).


```python
plt.figure(figsize=(10,6))
plt.hist(data['VClass'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Fuel Efficiency')
plt.xlabel('Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# Interpretation:
# The heatmap displays the correlation between different variables in the dataset.
# For example, if 'Vehicle_Weight' and 'Fuel_Efficiency' have a negative correlation,
# it means that heavier vehicles tend to be less fuel-efficient.
# Positive correlations (close to 1) indicate strong relationships, while values close to 0 indicate weak or no relationship.


```python
numeric_data = data.select_dtypes(include=[np.number])
correlation_matrix = numeric_data.corr()


plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Transport Dataset')
plt.show()
```
